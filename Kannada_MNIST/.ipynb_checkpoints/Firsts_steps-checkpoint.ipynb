```python
import tensorflow as tf
from tensorflow import keras
```

```python
#Loading FASHION MNIST data as a module
fashion_mnist = keras.datasets.fashion_mnist
#deploying the module
(X_train_full,y_train_full),(X_test,y_test) = fashion_mnist.load_data()
#We are using a 28pixel version for this data
print(X_train_full.shape,X_test.shape)
```

```python
#We divided by 255.0 bcs we want the data to be floats instead of integers
X_valid, X_train = X_train_full[:5000]/255.0, X_train_full[5000:]/255.0
y_valid, y_train = y_train_full[:5000], y_train_full[5000:]
```

```python
class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat", "Sandal", 
               "Shirt", "Sneaker", "Bag", "Ankle boot"]
```

#### Creating my first ANN model with Keras

```python
#We will build our model architecture
model = keras.models.Sequential()
#Flatten converts the input on 1D-vector. We have to specify the input dimensions 
model.add(keras.layers.Flatten(input_shape=[28,28]))
model.add(keras.layers.Dense(300, activation='relu'))
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))

#Also we can do this and save us some lines:

model2 = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[28,28]),
    keras.layers.Dense(300, activation='relu'),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])
```

```python
#We can see the number of parameters in the model in here
model.summary()
```

```python
# We can also see the initial parameters of the model with .get_layer and .get_weights 
weights, biases = model.get_layer('dense_3').get_weights()
weights
```

```python
#Compiling the model...
#That means to set the loss function, the optimizer and the metrics.
# We will use stochastic gradient descent
model.compile(loss= "sparse_categorical_crossentropy",
              optimizer="sgd",
              metrics=["accuracy"])
#------------------------------------------------------------------------------------------------------------------
#------------------------------------------------------------------------------------------------------------------
#Notice that we use sparse_categorical_crossentropy because we have sparse labels, if we would predict a binary label
#we should use only categorical_crossentropy.
#------------------------------------------------------------------------------------------------------------------
#------------------------------------------------------------------------------------------------------------------
```

```python
history = model.fit(X_train,y_train, epochs=30,
                     validation_data=(X_valid, y_valid))
```

## We can easily see the parameters of my model:

```python
import pandas as pd
import matplotlib.pyplot as plt
history.params
```

```python
#Also que can see the training parameters of each epochs. We can plot it with pandas easily.
pd.DataFrame(history.history).plot(figsize=(10,8))
plt.grid(True)
plt.gca().set_ylim(0,1) #set the vertical range to [0-1]
plt.show()
```

```python
#Validading the model with my test data
model.evaluate(X_test, y_test)
```

### Using the model to make a prediction.

```python
#We will use only the first three instances of my test data.
X_new = X_test[:3]
y_proba = model.predict(X_new)
y_proba
#We should see that our model estimates probabilities for one of my 9 instances. 
```

```python
y_proba.round()
```

```python
# We can also predict labels as classes.
import numpy as np
y_pred = model.predict_classes(X_new)
np.array(class_names)[y_pred]
```

### Non-sequential neural networks. Wide & Deep Neural networks

```python
input_ = keras.layers.Input(shape=X_train.shape[1:])
hidden1 = keras.layers.Dense(30, activation='relu')(input_)
hidden2 = keras.layers.Dense(30, activation='relu')(hidden1)
concat = keras.layers.Concatenate()([input_,hidden2])
output = keras.layers.Dense(1)(concat)
model = keras.models.Model(inputs=[input_], outputs=[output])
```

```python
model.compile(loss= "sparse_categorical_crossentropy",
              optimizer="sgd",
              metrics=["accuracy"])

model.fit(X_train,y_train,epochs=30,validation_data=(X_valid, y_valid))
```

### Dynamic Models Using the Subclassing API

```python
class WideAndDeepModel(keras.models.Model):
    def __init__(self,units=30,activation='relu',**kwargs): #we can specify units in here
        super().__init__(**kwargs)
        self.hidden1 = keras.layers.Dense(units, activation=activation)#the same # of neurons
        self.hidden2 = keras.layers.Dense(units, activation=activation)
        self.main_output = keras.layers.Dense(1) #Output #1
        self.aux_output = keras.layers.Dense(1) #Output #2

    def call(self, inputs):
        input_A, input_B = inputs 
        hidden1 = self.hidden1(input_B)
        hidden2 = self.hidden2(input_A)
        concat = keras.layers.concatenate([input_A,hidden2])
        main_output = self.main_output(concat)
        aux_output = self.aux_output(hidden2)
        return main_output, aux_output
```
