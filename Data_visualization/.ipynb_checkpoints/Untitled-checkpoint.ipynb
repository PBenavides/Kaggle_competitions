```python
import pandas as pd
import numpy as np
import seaborn as sns 

import matplotlib.pyplot as plt 
%matplotlib inline

from plotly.offline import iplot
import plotly as py
import plotly.tools as tls 

import cufflinks as cf 
```

```python
data = pd.read_csv('USD_PEN_2019.csv')
```

```python
pd.set_option('display.max_rows',None)
data
```

```python
data_fechas = data['Fecha'].apply(lambda x: pd.Series([i for i in reversed(x.split('.'))]))
data_fechas.columns = ['Año','Mes','dia']
```

```python
data = data.join(data_fechas, how='outer')
```

```python
data = data.reindex(index = data.index[::-1]).reset_index()
data.drop('index',axis=1, inplace=True)
```

```python
data = data.reset_index()
```

```python
#data.drop('index',axis=1,inplace=True)
data
```

```python
data.Último = data.Último.str.replace(',','.').astype('float32')
data.Apertura = data.Apertura.str.replace(',','.').astype('float32')
data.Máximo = data.Máximo.str.replace(',','.').astype('float32')
data.Mínimo = data.Mínimo.str.replace(',','.').astype('float32')
data['% var.'] = data['% var.'].str.replace(',','.')
```

```python
data.Mes = data.Mes.astype('int16')
```

```python
data.index = data.index.astype('int32')
```

```python
data.drop(['Fecha','% var.','Año','dia'],axis=1,inplace=True)
```

```python
py.offline.init_notebook_mode(connected=True)
```

```python
cf.go_offline()
```

```python
data.iplot(kind='heatmap',x='index',y='Mes')
```
